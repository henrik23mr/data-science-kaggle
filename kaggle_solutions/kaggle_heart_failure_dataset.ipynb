Data Preprocessing

```python
y = df['output']
data = df
X = data.drop(columns=['output'])
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=42)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)

X_train[:10], X_test[:10]
```

#### Models

LogisticRegression

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
predicted = model.predict(X_test)
predicted
```

```python
from sklearn.metrics import confusion_matrix

conf = confusion_matrix(y_test, predicted)
conf
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, predicted)
```

Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(X_train, y_train)
```

```python
predicted = model.predict(X_test)

predicted
```

```python
accuracy_score(y_test, predicted)
```

#### Bernoulli Naive Bayes

```python
from sklearn.naive_bayes import BernoulliNB

model = BernoulliNB()
model.fit(X_train, y_train)
```

```python
predicted = model.predict(X_test)
predicted
```

```python
accuracy_score(y_test, predicted)
```

#### Support Vector Machine

```python
from sklearn.svm import SVC

model = SVC()
model.fit(X_train, y_train)
```

```python
predicted = model.predict(X_test)
predicted
```

```python
accuracy_score(y_test, predicted)
```

#### Random Forest

using GridSearchCv to find

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

random_forest = RandomForestRegressor()

random_forest_params = {
                 'n_estimators': [20,30, 50, 80, 100, 150, 200],
                 'max_depth': [1,2,3,4],
                 'min_samples_split': [2,3,4]}

random_forest_cv_model = GridSearchCV(random_forest, random_forest_params, cv=10, n_jobs=-1)

random_forest_cv_model.fit(X_train, y_train)
```

```python
print('Best score for train_set:' + str(random_forest_cv_model.best_score_))
print('_______________________________________')
print(' ')
print('best n_estimators value: ' + str(random_forest_cv_model.best_params_['n_estimators']),
      '\nbest max_depth value: ' + str(random_forest_cv_model.best_params_['max_depth']),
      '\nbest min_samples: ' + str(random_forest_cv_model.best_params_['min_samples_split']))
```

```python
random_forest_reg = RandomForestRegressor(n_estimators=random_forest_cv_model.best_params_['n_estimators'], min_samples_split=random_forest_cv_model.best_params_['min_samples_split'],
                                          max_depth=random_forest_cv_model.best_params_['max_depth'])

random_forest_reg.fit(X_train, y_train)
```

```python
predicted = random_forest_reg.predict(X_test)
predicted
```

```python
accuracy_score(y_test, predicted.round())
```

#### K Nearest Neighbours

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier

kneighbors = KNeighborsClassifier(n_neighbors=1)

kneighbors.fit(X_train, y_train)
```

```python
predicted = kneighbors.predict(X_test)

predicted
```

```python
accuracy_score(y_test, predicted)
```

```python
error_rate = []

for i in range(1, 40):

    model = KNeighborsClassifier(n_neighbors = i)
    model.fit(X_train, y_train)
    pred_i = model.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))

plt.figure(figsize =(10, 6))
plt.plot(range(1, 40), error_rate, color ='blue',
                linestyle ='dashed', marker ='o',
         markerfacecolor ='red', markersize = 10)

plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
kneighbors = KNeighborsClassifier(n_neighbors=7)

kneighbors.fit(X_train, y_train)
```

```python
predicted = kneighbors.predict(X_test)
predicted
```

```python
accuracy_score(y_test, predicted)
```
